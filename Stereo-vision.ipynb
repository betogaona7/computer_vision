# Stereo vision

It is a way of getting depth (3D) information about a scene from two (or more) 2D images.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Read images in grayscale 
left_img = cv2.imread("images/imgL.png",0)
right_img = cv2.imread("images/imgR.png", 0)
```

```python
plt.imshow(left_img)
```

```python
stereo = cv2.StereoBM_create(numDisparities=64, blockSize=7)
disparity = stereo.compute(left_img, right_img)
```

```python
plt.imshow(disparity, 'gray')
plt.show()
```
